### apply()
Pandas.apply allow the users to pass a function and apply it on every single value of the Pandas series.

```python
df.head()
```

```python
def rate(x):
    if x>36000:
        return 'expensive'
    else:
        return 'Non-Expensive'
        
```

```python
df['new_col']=df['price'].apply(rate)
```

```python
df.loc[5:15,['price','company','new_col']]
```

### convert name of companies to uppercase and store in a new col

```python
def upper(x):
    return x.upper()
```

```python
df['new']=df['company'].apply(upper)
```

```python
df[['company','new']].head()
```

### Anonymous function:


```python
x=lambda x,y:x+y
```

```python
x(1,2)
```

```python
df['gprice']=df['price'].apply(lambda x:x>17000)
```

```python
import numpy as np
company=['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
profit=[40, 2, 34, 12]
xpos=np.arange(len(company)) # xpos = [0,1,2,3]
plt.xticks(xpos, company) #yticks
plt.ylabel('revenue')
plt.title('US Tech Stocks')
plt.bar(xpos-0.2 ,revenue, label='Revenue', width=0.3)
plt.bar(xpos+0.2, profit, label='Profit',width=0.3)
plt.legend()
plt.show()
```
